```python
import pandas as pd

# archivo
archivoMaterias = "db/Carreras Materias.xlsx"
archivoPA = "db/pregrado/Pregrado PA 2025.xlsx"
archivoPO = "db/posgrado/Posgrados PA 2025.xlsx"

materias = pd.read_excel(archivoMaterias, sheet_name="Materias")
pa = pd.read_excel(archivoPA, sheet_name="Data")
po = pd.read_excel(archivoPO, sheet_name="Data")
```

```python
# Extraccion nombres a comprobar
nombreMaterias = materias["MATERIA"]

# Recorrer lista nombre materias
for materia in nombreMaterias:
    filtro = pa["MATERIA_TITULO_PA"] == materia
    dFiltrado = pa[filtro]

    semestres = dFiltrado["PERIODO"].unique()

    # Mostrar materias que no estan en todos los semestres
    if (202510 not in semestres or 202520 not in semestres) and semestres.size != 0:
        print(f"La materia: {materia} esta en {semestres}")
```

```python
# filtros para Programacion Academica
filtroFacultad = pa["FACULTAD"] == "FAC. DE CIENCIAS DE LA SALUD"
filtroLaboratorio = pa["SSBSECT_SCHD_CODE"] == "PRA"

paFiltrado = pa[filtroFacultad]
paFiltrado = paFiltrado[filtroLaboratorio]

materiasSemestre = []

# Recorrer lista nombre materias
for materia in paFiltrado["MATERIA_TITULO_PA"].unique():
    for materiaS in materiasSemestre:
        if(materiaS == materia):
            siglas = paFiltrado[paFiltrado["MATERIA_TITULO_PA"] == materia]["SIGLA"].unique()
            print(f"{materia}: {siglas}")
```

```python
siglas_objetivo = [
    "ECAR0017",
    "ECAR0018",
    "ECAR0019",
    "ECAR0020",
    "ECAR0021",
    "ECAR0022",
    "TTMZ0753",
    "ECAR0023",
]

# Verificar existencia de registros por sigla
print("üìå Revisi√≥n de existencia de registros por sigla:")
for sigla in siglas_objetivo:
    if sigla in po["SIGLA"].values:
        print(f"‚úÖ La sigla {sigla} s√≠ tiene registros.")
    else:
        print(f"‚ùå La sigla {sigla} no tiene registros en el archivo.")

# Aplicar los filtros
po_filtrado = po[po["SIGLA"].isin(siglas_objetivo)]
po_filtrado = po_filtrado[po_filtrado["SSRMEET_SCHD_CODE"] == "PRA"]
po_filtrado = po_filtrado[po_filtrado["SALA"].notna() & (po_filtrado["SALA"] != "/")]

# Mostrar resultados que cumplen las condiciones
print("\nüìå Materias que cumplen con PRA y SALA v√°lida:")
for sigla in siglas_objetivo:
    materias = po_filtrado[po_filtrado["SIGLA"] == sigla]["MATERIA_TITULO"].unique()
    if len(materias) > 0:
        print(f"‚úÖ Sigla: {sigla} ‚Üí Materia(s): {materias}")
    else:
        print(f"‚ö†Ô∏è  Sigla: {sigla} no cumple con PRA y/o SALA v√°lida.")
```

```python
def aplicar_tonalidad(hex_color, ocup):
    """
    Mezcla el color base con blanco seg√∫n ocup (0.0‚Äì1.0).
    ocup=0 ‚Üí color = blanco
    ocup=1 ‚Üí color = base
    """
    # Limpieza y parseo
    hex_color = hex_color.lstrip("#")
    try:
        r, g, b = (int(hex_color[i : i + 2], 16) for i in (0, 2, 4))
    except ValueError:
        # Fallback: gris claro
        return "#CCCCCC"

    # Mezcla lineal con blanco (255,255,255)
    r_new = int(r * ocup + 255 * (1 - ocup))
    g_new = int(g * ocup + 255 * (1 - ocup))
    b_new = int(b * ocup + 255 * (1 - ocup))

    return f"#{r_new:02X}{g_new:02X}{b_new:02X}"


for o in [0.0, 0.25, 0.5, 0.75, 1.0]:
    print(o, aplicar_tonalidad("#1F77B4", o))
```
